# 2) data Visualization

    graphing the features in a pair plot, it is clear that the relationship between pairs of features of a iris-setosa (in pink) is distinctly different from those of the other two species. For no of feature mpore than two then we use pairplot

```python

iris_data = Iris.drop('Id', axis=1)
sns.pairplot(iris_data, hue= 'Species')
plt.show()
```

# 3) Decision Tree Model

```python
# Import label encoder
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

# Encode text labal into number
iris_data['Species'] = label_encoder.fit_transform(iris_data['Species'])

x = iris_data.iloc[:,0:4].values
y = iris_data.iloc[:,4].values
print(y)
```

```python
# imorting decision tree library
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
Tree_model = DecisionTreeClassifier()
Tree_model.fit(x,y)
print(" Training Completed")
```

# 4) Decision Tree Visualization

```python
# tree visulization using sklearn module
iris_feature = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
iris_class = ['Iris-versicolor','Iris-virginica','Iris-setosa ']
fig = plt.figure(figsize = (15,10))

tree.plot_tree(Tree_model, feature_names = iris_feature,
              class_names = iris_class,filled = True);
plt.show()
```
